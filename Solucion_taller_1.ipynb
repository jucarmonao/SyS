### **Ejercicio** 3.1.1
* Graficar la potencia media de una impedancia  R+jwL , cuándo es excitada por una fuente sinusoidal.
Esto nos determina un circuito RL por tanto procedemos a sacar las ecuaciones de cada elemento las cuales son:
$$R(t) = R i(t)$$
$$v_L(t) = L\frac{d}{dt}i(t)$$

De modo que

$$
v_R(t) = R I_{max}\sin(\omega t)
$$
$$
v_L(t) = L \frac{d}{dt}\bigg(I_{max}\sin(\omega t)\bigg) = \omega L I_{max}\cos(\omega t)
$$

Por leyes de Kirchhoff sabemos que

$$
v(t) = v_R(t) + v_L(t)\\
v(t) = R I_{max}\sin(\omega t) + \omega L I_{max}\cos(\omega t)
$$

Desde un inicio se había dicho que $v(t) = V_{max}\sin(\omega t + \theta)$, así que comparando ambas expresiones se puede determinar los valores de $V_{max}$ y $\theta$

$$
V_{max}\sin(\omega t + \theta) = R I_{max}\sin(\omega t) + \omega L I_{max}\cos(\omega t)\\
V_{max}\big(\sin(\omega t)\cos(\theta)+\sin(\theta)\cos(\omega t)\big) = R I_{max}\sin(\omega t) + \omega L I_{max}\cos(\omega t) 
$$

Comparando los términos se pueden establecer las siguientes igualdades:

$$
V_{max}\cos(\theta) = RI_{max}\\
V_{max}\sin(\theta) = \omega L I_{max}
$$

Dividiendo la segunda entre la primera se puede llegar al valor de $\theta$

$$
\theta = \arctan\bigg(\frac{\omega L}{R}\bigg)
$$

Usando la identidad trigonométrica pitagórica se obtiene el valor para $V_{max}$

$$
V_{max} = \sqrt{R^2 + (\omega L)^2} I_{max}
$$

Ahora podemos determinar la tensión a partir de la corriente para un circuito serie RL

Determinemos su potencia media a lo largo de un periodo, recordando que la potencia media se define como

$$
P = \frac{1}{T_0} \int_0^{T_0} p(t)~dt
$$

Para una Carga $Z=R+jwL$ si $p(t) = Z_{mag}*i(t)^{2}$ entonces 


$$
p(t)=sen(wt)^2(R+jwL)
$$
$$
\bar{P}i = \frac{1}{T} \int_0^{T} Z{mag}* i(t)^{2}~dt= \frac{1}{T} Z\int_0^{T} i(t)^{2}~dt 
$$

$$

#### 5.3

Sea la señal Gaussiana $x(t) = e^{−at2}$ con $a \in R^{+}$,el sistema A con relación entrada-salida $y_{A}(t) = x^{2}(t)$, y el sistema lineal e invariante con el tiempo B con respuesta al impulso $h(t) = Be−bt2:$  

* a) Encuentre la salida del sistema en serie $x(t) → h(t) → y_{A}(t) → y(t)$
Al tener la respuesta al escalón se realizara la convolución entre las señal $x(t)$ y la respuesta $h_{B}(t)$, así:

$x(t)* h_{B}(t)=\int_{-\infty}^{\infty}(Be^{-bT^{2}}e^{-a(t-T)^{2}})dT$

$\int_{-\infty}^{\infty}(Be^{-bT^{2}-a(t-T)^{2}})dT=$$\frac{\sqrt{\pi}Be^{-\frac{abt^{2}}{b+a}}}{\sqrt{b+a}}$


En la continuación del sistema se aplica a la salida $y_{A}(t)=x^{2}(t)$, y asi resolviendo esto tenemos 
$y_{A}(t)=x^{2}(t)=(\frac{\sqrt{\pi}Be^{-\frac{abt^{2}}{b+a}}}{\sqrt{b+a}})^{2}=\frac{\pi B^{2}e^{-2(\frac{abt^{2}}{b+a})}}{b+a}=\frac{\pi B^{2}e^{(2at^{2})(\frac{a}{a+b}- 1)}}{b+a}=y(t)$

Simulamos en sympy:

```python
from sympy import *
import sympy as sym
sym.init_printing()
a,x,b,t,B = symbols('a x b t B', positive = True) 

from sympy import oo, Symbol, integrate
def convolve(f, g, t, lower_limit=-oo, upper_limit=oo):
    tau = Symbol('tau', real=True)
    return integrate(f.subs(t, tau) * g.subs(t, t - tau), 
                     (tau, lower_limit, upper_limit))

ft = exp(-a*t**2)
gt = B*exp(-b*t**2)
simplify(convolve(ft,gt,t)**2)
```

b) Encuentre la salida del sistema en serie $x(t) → y_{A}(t) → h(t) → y(t)$. 

Aplicamos la salida  $y_{A}(t)=x^{2}(t)$


$x(t)^{2} = e^{2(-at^{2})}$


Ahora realizamos la convolución de los siguientes $x(t)^{2} * h_{B}$


$x(t)^{2} * h_{B} = \int_{-\infty}^{\infty}(Be^{-bT^{2}}  e^{-2a(t-T)^{2}} )dT = \frac{\sqrt{\pi} B e^{2at^{2}\left(\frac{2a}{2a+b} - 1 \right)}}{\sqrt{2a + b}} = y(t)$

```python
from sympy import *
import sympy as sym
sym.init_printing()
a,x,b,t,B = symbols('a x b t B', positive = True) 

from sympy import oo, Symbol, integrate
def convolve(f, g, t, lower_limit=-oo, upper_limit=oo):
    tau = Symbol('tau', real=True)
    return integrate(f.subs(t, tau) * g.subs(t, t - tau), 
                     (tau, lower_limit, upper_limit))

ft = exp(2*-a*t**2)
gt = B*exp(-b*t**2)
simplify(convolve(ft,gt,t))
```
