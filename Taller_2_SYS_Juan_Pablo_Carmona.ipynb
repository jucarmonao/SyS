##2.2. Demuestre la ortogonalidad del conjunto $\{cos(nω_ot),sin(mω_ot)\}_{n,m∈\mathbb{Z}}$

Tenemos: 

\begin{equation}
cos(nw_ot)cos^*(mw_ot)
\end{equation}

Entonces:

\begin{equation}
 \int_{-T/2}^{T/2}cos(nw_ot)cos^*(mw_ot)dt
\end{equation}

Si n=m, entonces:

\begin{equation}
\int_{-T/2}^{T/2}cos(nw_ot)cos(nw_ot)dt 
\end{equation}

\begin{equation}
 \int_{-T/2}^{T/2}cos^2(nw_ot)dt
\end{equation}

Aplicando la identidad:

\begin{equation}
cos^2(\theta)= \dfrac{1}{2} + \dfrac{cos(2\theta)}{2}, \  cos (\theta=nw_ot)
\end{equation}

Tenemos:

\begin{equation}
\int_{-T/2}^{T/2}\dfrac{1}{2} + \dfrac{cos(2nw_ot)}{2}
\end{equation}

Integrando tenemos:

\begin{equation}
\dfrac{T}{2}+\dfrac{1}{4nw_o}[sin(2n\pi)-sin(-2n\pi)]
\end{equation}

Como $Sin(2n\pi)=0$ entonces:

\begin{equation}
E_c=\dfrac{T}{2}
\end{equation}

Si son diferentes entonces:

\begin{equation}
  \int_{-T/2}^{T/2}cos(mw_ot)cos(nw_ot)dt
\end{equation}

Teniendo en cuenta lo siguiente:

\begin{equation}
 cos(\alpha + \beta)= cos(\alpha)cos(\beta) - sin(\alpha)sin(\beta)
\end{equation}

\begin{equation}
  cos(\alpha - \beta)= cos(\alpha)cos(\beta) + sin(\alpha)sin(\beta)
\end{equation}

Tendriamos:

\begin{equation}
  cos(\alpha +\beta) + cos(\alpha - \beta)= 2cos(\alpha)cos(\beta)
\end{equation}

\begin{equation}
  cos(\alpha)cos(\beta)= \dfrac{1}{2}(cos(\alpha-\beta)+cos(\alpha+\beta))
\end{equation}

Usando esto en la integral obtenemos:

\begin{equation}
  \dfrac{1}{2}\int_{-T/2}^{T/2}cos(nw_ot-mw_ot)+cos(nw_ot+mw_ot)dt
\end{equation}

\begin{equation}
  \dfrac{1}{2}\int_{-T/2}^{T/2}cos((n-m)w_ot)+cos((n+m)w_ot)dt
\end{equation}

Sustituimos $n-m=r_1$ y $n+m=r_2$ y $w_o= \frac{2\pi}{T}$ y resolvemos:

\begin{equation}
 \dfrac{1}{2}(\dfrac{sin(r_1\pi)+sin(-r_1\pi)}{r_1w_o}-\dfrac{sin(r_2\pi)+sin(-r_2\pi)}{r_2w_o}) = 0
\end{equation}

 \begin{equation}
  E_c=0; \quad r_1,r_2 \in \mathbb{Z}
 \end{equation}

```
from sympy import *
t = Symbol("t")
m = Symbol("m", integer=True,positive=True)
n = Symbol("n", integer=True,positive=True)
T = Symbol("T",positive=True)
Wo = 2*pi/T
integrate(cos(n*Wo*t)*cos(m*Wo*t),(t,-T/2,T/2))
```

 \begin{equation}

La representacion por polinomios de legendre consiste en de cierta forma, poder recrear ciertas ondas de  una forma más sencilla al momento de hacer operaciones con ella, utilizada en muchos casos para hallar las integrales definidias, e infinidad de aplicaciones que se ha encontrado.

Las funciones de legendre provienen de la solución de ecuaciones diferenciales de legendre, las cuales son:
$$ 
\frac{d}{dx}[(1-x^{2})\frac{d}{dx}P_{n}(x)]+n(n+1)p_{n}(x)=0
$$

Los Polinomios de Legendre son uno de los ejemplos más importantes de los Polinomios Ortogonales, porque
aparecen como soluciones en varios problemas clásicos de la física como:
- Movimiento de los planetas: Ecuación de Kepler.
- Resolución de los modelos de la física con Ecuaciones Diferenciales en Derivadas Parciales (EDDP) en coordenadas esféricas.

Son ejemplo de estos modelos de la física, los campos conservativos, no conservativos, propagación de calor,
propagación de ondas, propagación de señales telegráficas, propagación de ondas de partículas simples. 
## Análisis de la ortogonalidad de las bases de Legendre

El grupo de polinomios de legendre $P_n(x)$, con $n \in \mathbb{Z}^+$ forma un conjunto completo de funciones mutuamente ortogonales en el intervalo $(-1 < t < 1)$ y se define como:

$$
P_n(t) = \frac{1}{2^nn!}\frac{d^n}{dt^n}\left\{(t^2-1)^n\right\},
$$

Por ende:

$$
\begin{align}
P_0(t) =& 1\\
P_1(t) =& t\\
P_2(t) =& \left(\frac{3}{2}t^2-\frac{1}{2}\right)\\
P_3(t) =& \left(\frac{5}{2}t^3-\frac{3}{2}\right)\\
\vdots =& \vdots
\end{align}
$$

Se puede verificar la ortogonalidad de estos polinomios (obviando el $p_{0}$) igual a 1.



```
from sympy import *
init_printing()
t = symbols('t', real=True)

for i in range(1, 6):
    print("P{}(t) =".format(i))
    pprint(legendre(i, t))
    print(end=2*'\n')
```

```
import matplotlib.pyplot as plt
import numpy as np

tn = np.linspace(-1, 1, 1000)
plt.figure(figsize=(10, 8))
plt.plot(tn, np.ones(tn.shape), label="$P_0(t)$")
for i in range(1, 6):
    plt.plot(tn, lambdify(t, legendre(i, t), "numpy")(tn), label="$P_{}(t)$".format(i))
plt.legend(loc="best", fontsize=12)
plt.xlabel("Tiempo")
plt.title("Polinomios de Legendre")
plt.grid()
```

Definimos la función que usaremos en cada parte 

```
def Leg_Serie(x, N):
    C = []
    xl = 0
    for degree in range(N):
        C.append((integrate(x*legendre(degree, t), (t, -1, 1)))/(integrate(legendre(degree, t) ** 2, (t, -1, 1))))
        xl += C[degree] * legendre(degree, t)
    return C, xl
```

Caso p1 = t

```
tn = np.linspace(-1, 1, 1000)
x =  legendre(1, t)

C, xl = Leg_Serie(x, 2)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
ax1.plot(tn, lambdify(t,  x, "numpy")(tn), label="Señal original")
ax1.plot(tn, lambdify(t,  xl, "numpy")(tn), label= "Señal reconstruida")
ax1.grid()
ax1.legend(loc="best")
ax1.margins(x=0)
ax1.set_ylim(-1.5, 1.5)
ax1.set_xlabel("Tiempo")
ax1.set_ylabel("Señales")

ax2.stem(np.arange(len(C)), [i.evalf() for i in C], use_line_collection=True)
ax2.grid()
ax2.set_xlabel("Grado")
ax2.set_ylabel("Valor coeficiente")

fig.show()
```

**3.SERIE DE FOURIER**

Antes de empezar con los puntos debemos definir la funcion de fourier puesto que sera necesaria en multiples ocasiones a lo largo del punto.

```
def SF(T,fs,xo,N,cn,co,bandera,px):
    Nm = T*Fs
    tv = np.arange(-T/2,T/2,T/Nm)
    xo = xo(tv)

 
 
    wo = 2*np.pi/T # frecuencia fundamentas
    # %% definir bases
    print(Nm)
    phin = np.zeros((len(tv),2*N+1),dtype=np.complex_)
    for n in range(-N,N+1,1):
        phin[:,n+N] = np.exp(1j*n*wo*tv)
   

    nv = np.linspace(-N, N, 2*N+1) #vector num armónicos
    Cn = cn(nv)
    Cn[N] = co
    if bandera == 1:
        Cn[N-1] = 0
        Cn[N+1] = 0
    if bandera == 2:
        Cn[N+1] = 1/2
       

    fig = plt.figure(figsize=(8,15))
    plt.subplot(7,1,1)  
    plt.plot(tv,xo,'r',linewidth=4)
    plt.grid()
    plt.xlabel("t[s]",fontsize = 14)
    plt.ylabel("x(t)",fontsize = 14)

    plt.subplot(7,1,2)
    plt.stem(nv, np.real(Cn), 'r')
    plt.title(r'Re{Cn}')
    plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
    plt.ylabel(r'$Re\{C_n\}$',fontsize = 14)
    plt.axis('tight')
    plt.grid()

    plt.subplot(7,1,3)
    plt.stem(nv, np.imag(Cn), 'r')
    plt.title(r'Im{Cn}')
    plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
    plt.ylabel(r'$Im\{C_n\}$',fontsize = 14)
    plt.axis('tight')
    plt.grid()

    plt.subplot(7,1,4)  
    plt.stem(nv, abs(Cn), 'r')
    plt.title(r'|Cn|')
    plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
    plt.ylabel(r'$|C_n|$',fontsize = 14)
    plt.axis('tight')
    plt.grid()

    plt.subplot(7,1,5)  
    plt.stem(nv, np.arctan(np.imag(Cn)/np.real(Cn)), 'r')
    plt.title(r'|Cn|')
    plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
    plt.ylabel(r'$Espectro$',fontsize = 14)
    plt.axis('tight')
    plt.grid()
 

    plt.subplot(7,1,6)  
    xe = phin @ Cn  #producto matricial para calcular componentes espectrales
    plt.plot(tv,xe)
    plt.title("x^(t)")
    fig.tight_layout()
    Na = 10
    ind = range(N-Na,N+Na+1)
    er = 1 - (np.sum(abs(Cn)**2))/px
 
   
    xe = phin[:,ind] @ Cn[ind]
    plt.subplot(7,1,7)    
    plt.plot(tv,xe,color='b',label='xe')
    plt.plot(tv,xo,color='r',label='x')
    plt.plot(tv,xo-xe,color='g',label='Error')
    plt.title('Er=%.2f[%%]' % (100*er),fontsize=16)
    plt.xlabel('t[s]')
    plt.ylabel('x(t)')
    plt.grid()
    plt.legend()
    plt.show()
    if bandera == 3:
        print("Error cuadrático medio: " + str(round(er,4)))
```

\begin{equation}

3.2) Sea $x^{''}(t)$ la segunda derivada de la señal $x(t)$, donde $t\in [t_i,t_f]$. Demuestre que los coeficientes de la serie exponencial de Fourier se pueden calcular según$\\$
 $C_{n}=\frac{1}{(t_i-t_f)n^{2}\omega_{o}^{2}}\int_{ti}^{tf}x^{''}(t)e^{-jn\omega_{o}t}dt;n \in \mathbb{Z}$

Primero hallamos $C_n$ usando la formula general:

\begin{equation}
    C_n=\dfrac{1}{T}\int_{-T/2}^{T/2}x(t)e^{-jnw_ot}dt
\end{equation}


Integración por partes:

\begin{equation}
    u=x(t); dv=e^{-jnw_ot}
\end{equation}

\begin{equation}
    C_n= \dfrac{1}{T}[x(t)\dfrac{e^{-jnw_ot}}{-jnw_o}|{-T/2}^{T/2}+\int{-T/2}^{T/2}\dfrac{x'(t)e^{-jnw_ot}}{jnw_o} dt]
\end{equation}

Nuevamente integramos por partes:

\begin{equation}
    u=x'(t); dv=e^{-jnw_ot}

**SERIE DE FOURIER:** es una serie infinita que converge puntualmente a una función periódica y continua a trozos (o por partes). Las series de Fourier constituyen la herramienta matemática básica del análisis de Fourier empleado para analizar funciones periódicas a través de la descomposición de dicha función en una suma infinita de funciones sinusoidales mucho más simples, la cual esta definida de la siguiente manera:
Si${\displaystyle f(t)\,}$ es una función de variable real ${\displaystyle t\,}$, que es integrable en el intervalo ${\displaystyle [t_{0}-T/2,t_{0}+T/2]\,}$, entonces se puede obtener el desarrollo en serie de Fourier de ${\displaystyle f\,}$, en ese intervalo. Fuera del intervalo la serie es periódica, con período ${\displaystyle T\,}$, Si ${\displaystyle f(t)\,}$ es periódica en toda la recta real, la aproximación por series de Fourier también será válida en todos los valores de $t\$ 

Luego la serie de Fourier asociada a ${\displaystyle f(t)\,}$ es:

\begin{equation}
    \frac{a_0}{2}+\sum_{n=1}^{\infty}[a_n\cos\frac{2n\pi}{T}t+b_n\sin\frac{2n\pi}{T}t]
\end{equation}

Donde $a_n$ y $b_n$ son coeficientes de fourier

Las formulas que rigen la serie de fourier para la forma trigonometrica son:

\begin{equation}
    f(t)=\frac{a_0}{2}+\sum_{n=1}^{\infty}(a_n\cos\omega_nt+b_n\sin\omega_nt)
\end{equation}

\begin{equation}
    a_0=\frac{2}{T}\int_{t_0}^{t_0+T}f(t)dt
\end{equation}

\begin{equation}
    a_n=\frac{2}{T}\int_{t_0}^{t_0+T}f(t)\cos(n\omega_nt)dt
\end{equation}

\begin{equation}
    b_n=\frac{2}{T}\int_{t_0}^{t_0+T}f(t)\sin(n\omega_nt)dt
\end{equation}

Las formulas que rigen la serie de fourier para la forma compacta son:

\begin{equation}
    f(t)=A_0+\sum_{n=1}^{\infty}(A_n\cos(\omega_nt-\theta_n)
\end{equation}


\begin{equation}
    A_0=\frac{a_0}{2}
\end{equation}


\begin{equation}
    A_n=\sqrt{a_n^2+b_n^2}
\end{equation}

\begin{equation}
    \theta_n=\tan^{-1}\frac{b_n}{a_n}
\end{equation}

Las formulas que rigen la serie de fourier para la forma exponencial son:

\begin{equation}
    f(t)=\sum_{n=-\infty}^{\infty}c_ne^{j\omega_nt}
\end{equation}

\begin{equation}
    c_0=\frac{1}{T}\int_{t_0}^{t_0+T}f(t)dt \rightarrow \frac{a_0}{2}
\end{equation}

\begin{equation}
    c_n=\frac{1}{T}\int_{t_0}^{t_0+T}f(t)e^{-j\omega_nt}dt \rightarrow \frac{1}{2}(a_n-jb_n)
\end{equation}

\begin{equation}
    c_{-n}=\frac{1}{T}\int_{t_0}^{t_0+T}f(t)e^{j\omega_nt}dt \rightarrow \frac{1}{2}(a_n+jb_n)\\
\end{equation}

\begin{equation}
    |c_n|=\frac{1}{2}\sqrt{a_n^2+b_n^2}
\end{equation}

**TRANSFORMADA DE FOURIER:** Denominada así por Joseph Fourier, es una transformación matemática empleada para transformar señales entre el dominio del tiempo (o espacial) y el dominio de la frecuencia, que tiene muchas aplicaciones en la física y la ingeniería.   
Sea $x(n)$ una señal aperiodica y discreta en el tiempo. La transformada de Fourier discreta, se define como:

\begin{equation}
x(k)=\sum_{n=0}^{N-1}x(n)e^{-j\frac{2\pi n}{N}k}
\end{equation}

Uno de los principales métodos para realizar el cambio de DFT a FFT es aprovechando la simetría y dividiendola a un par más pequeño:

\begin{equation}
    x(k)=\sum_{n=0}^{N-1}x(n)e^{\frac{-j2\pi kn}{N}}
\end{equation}

\begin{equation}
=\sum_{m=0}^{\frac{N}{2}-1}x(2m)e^{\frac{-j2\pi kn}{N/2}}+\sum_{m=0}^{\frac{N}{2}-1}x(2m)e^{\frac{-j2\pi kn}{N/2}}
\end{equation}

Ordenando la información anterior, y agrupandola podemos verla mucho mejor


SERIE DE FOURIER |TRANSFORMADA DE FOURIER
:--------: | :-------:
Descomponen una función periódica en una suma de senos y cosenos con diferentes frecuencias y amplitudes. |Operación matemática que rompe una señal en sus frecuencias constituyentes.
Expansión de la señal periódica como una combinación lineal de senos y cosenos. | Representación en el dominio de la frecuencia de una señal, ya que depende de la frecuencia.

$\hat{x}(t)=\sum_{n=-\infty}^{\infty}C_{n}e^{j\omega_{o}t}$ | $\hat{x}(t)=\frac{1}{2\pi}\int_{-\infty}^{\infty}x(\omega)e^{j\omega t}d\omega \rightarrow x(t)=F^{-1}${$x(\omega)$}

```
def TF(T,Fs,xo):
    Nm = T*Fs
    tv = np.arange(-T/2,T/2,T/Nm)
    xo = xo(tv)
    plt.title("Señal Original")
    plt.plot(tv,xo,"b")
    plt.show()
    vf = np.fft.fftfreq(len(xo),1/Fs) #calculo vector de frecuencias
    Xw = np.fft.fft(xo) #cálculo FFT
    plt.stem(vf,abs(Xw),'r')
    plt.title(r'|X[n]|')
    plt.xlabel(r'$f$[Hz]',fontsize = 14)
    plt.ylabel(r'$|X[n]|$',fontsize = 14)
    plt.axis('tight')
    plt.grid()
    plt.show()
    plt.stem(vf,abs(Xw)/len(vf),'r')
    plt.title(r'|X[n]|/K=c_n')
    plt.xlabel(r'$f$[Hz]',fontsize = 14)
    plt.ylabel(r'$|X[n]|/K$',fontsize = 14)
    plt.show()
    
    xe = np.fft.ifft(Xw)
    plt.plot(tv,xe)
    plt.title(r'Reconstrucción completa')
    plt.xlabel(r't',fontsize = 14)
    plt.ylabel(r'$x[n]$',fontsize = 14)
    plt.show()
    ind = range(10,1000)
    Xwf = Xw.copy()
    Xwf[ind] = 0
    xe2 = np.fft.ifft(Xwf)
    plt.title(r'Señal ponderada')
    plt.plot(tv,xe2)
    plt.show()
```

**LA DISTORCION ARMÓNICA TOTAL ( THD ):** es una medida de la distorsión armónica presente en una señal y se define como la relación de la suma de las potencias de todos los componentes armónicos a la potencia de la frecuencia fundamental , que se puede traducir en:

\begin{equation}
    THD=\sqrt{D_2^2+D_3^2+.....}
\end{equation}

**DISTORCIÓN ARMONICA:**la alteración de la forma de onda de una portadora de información de señal , tal como una señal de audio que representa un sonido o señal de vídeo que representa imágenes, en un dispositivo o electrónica canal de comunicación . La distorsión es por lo general no deseado, y para que los ingenieros se esfuerzan por eliminar o minimizar ella. En algunas situaciones, sin embargo, la distorsión puede ser deseable. Por ejemplo, en la radiodifusión de FM y de reducción de ruido sistemas como el sistema Dolby, la alteración de la forma de onda de una portadora de información de señal , tal como una señal de audio que representa un sonido o señal de vídeo que representa imágenes, en un dispositivo o electrónica canal de comunicación.

El factor de potencia determina la eficiencia de su consumo eléctrico, a la hora de convertirlo en potencia útil, como luz, calor o movimiento mecánico y es la relación entre la potencia eficaz y la potencia aparente total:



\begin{equation}
\mbox{Potencia Aparente S} = VIrms = \sqrt{I_1^{2} + I_2^{2}+...+I_n^{2}}
\end{equation}


\begin{equation}
\mbox{Potencia Activa P} = VI*cos(\varphi_1)
\end{equation}


\begin{equation}
\mbox{Factor de potencia PF} = \frac{P}{S}
\end{equation}
